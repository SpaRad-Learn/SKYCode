```python
ratings
```

```python
ratings['Name']
```

```python
ratings['Name']= ratings['Name'].str.replace(r'\W', '',regex=True)
```

```python
ratings['Name']
```

```python
ratings['Domain']= ratings['Domain'].str.replace(r'\W', '',regex=True)
```

```python
ratings['Domain']
```

```python
ratings['Salary']= ratings['Salary'].str.replace(r'\W', '',regex=True)
```

```python
ratings['Salary']
```

```python
ratings['Location']= ratings['Location'].str.replace(r'\W', '',regex=True)
```

```python
ratings['Location']
```

```python
ratings['Age']= ratings['Age'].str.replace(r'\W', '',regex=True)
```

```python
ratings['Age']
```

```python
ratings['Age']= ratings['Age'].str.extract('(\d+)')
```

```python
ratings['Age']
```

```python
ratings['Exp']= ratings['Exp'].str.extract('(\d+)')
```

```python
ratings
```

```python
clean_data = ratings.copy()
```

```python
clean_data
```

```python
clean_data.isnull()
```

```python
import numpy as np
```

```python
clean_data['Age'] = clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data['Age'])))
```

```python
clean_data['Exp'] = clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp'])))
```

```python
clean_data['Exp']
```

```python
clean_data['Location']= clean_data['Location'].fillna(clean_data['Location'].mode()[0])
```
